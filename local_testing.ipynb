```python

#test
image_folder = "F:\dataset\ISIC2018\ISIC2018_Task3_Test_Input"  # Update with the correct path
output_folder = "F:\dataset\ISIC2018\ISIC2018_test"  # Where to save organized dataset
```

```python
# Create output subfolders based on class names (columns in the CSV except 'image')
class_label = df.columns[1:]
for label in class_label:
    os.makedirs(os.path.join(output_folder, label), exist_ok = True)
    
```

```python
for index, row in df.iterrows():
    image_name = row['image']+ '.jpg'
    # for label in class_label:
    #     if row[label] == 1:
    #         shutil.copy(os.path.join(image_folder, image_file), os.path.join(output_folder, label, image_file))

    

    image_path = os.path.join(image_folder, image_name)
    # print(image_path)
    # break

    if not os.path.exists(image_path):
        print(f"Image {image_name} not found! Skipping...")
        continue

    label = class_label[row[1:].tolist().index(1)]  # Convert to list and find the index of 1
    # print(label)
    destination_folder = os.path.join(output_folder, label)
    destination_path = os.path.join(destination_folder, image_name)

    #copy the image to the destination folder
    shutil.copy(image_path, destination_path)
    print(f"Image {image_name} copied to {destination_path}")

print("Done!")
    
```
